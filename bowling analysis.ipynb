```python
# This user define function will give the type,count of null and non null values as well as null ratio
def playerinfo():
  temp=pd.DataFrame(index=deliveries.columns)
  temp["datatype"]=deliveries.dtypes
  temp["not null values"]=deliveries.count()
  temp["null value"]=deliveries.isnull().sum()
  temp["% of the null value"]=deliveries.isnull().mean()
  temp["unique count"]=deliveries.nunique()
  return temp
playerinfo()
```

```python
def is_bowlers_wicket(player_dismissed,dismissal_kind):
    if type(player_dismissed)==str:
        if dismissal_kind not in ['run out','retired hurt','obstructing the field']:
            return 1
        else:
            return 0
    else:
        return 0
```

```python
deliveries['isBowlersWkt'] = deliveries.apply(lambda x: is_bowlers_wicket(x['player_dismissed'], x['dismissal_kind']), axis=1)
```
