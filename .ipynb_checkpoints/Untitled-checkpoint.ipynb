##Calculate the total runs conceded by each bowler. This will include the runs scored by the batsman (batsman_runs) and any extras (extras), except for byes and leg byes since they are not attributed to the bowler.
Count the distinct number of matches (matchId) each bowler has played.

```python
# Perform the bowling analysis to calculate total runs conceded and number of matches played by each bowler

# Step 1: Calculate total runs conceded by each bowler
# Note: Total runs conceded for a delivery = 'batsman_runs' + 'extras' - 'Byes' - 'LegByes'
# Replace NaN values with 0 for calculation
df['Byes'] = df['Byes'].fillna(0)
df['LegByes'] = df['LegByes'].fillna(0)
df['Total_Runs_Conceded'] = df['batsman_runs'] + df['extras'] - df['Byes'] - df['LegByes']

# Step 2: Group by bowler to calculate the total runs and count distinct matches
bowler_stats = df.groupby('bowler').agg(
    Total_Runs_Conceded=pd.NamedAgg(column='Total_Runs_Conceded', aggfunc='sum'),
    Matches_Played=pd.NamedAgg(column='matchId', aggfunc=pd.Series.nunique)
).reset_index()

bowler_stats.head()
```

```python
# Find the top 10 bowlers who have conceded the most runs along with the number of matches they played
top_10_bowlers = bowler_stats.sort_values(by='Total_Runs_Conceded', ascending=False).head(10)
top_10_bowlers


```

```python
## Lets do the bowling analysis of the below :
## R Vinay Kumar, Mohammad Siraj, KV Sharma , Akash Deep, GJ Maxwell, MK Lomror, MAyank Dagar, RJW Topley, C Green, AS Joseph, Yash Dayal, TK Curran, LH  Ferguson
```

```python

# Filtering to ensure we only include the bowlers actually present in the analysis
bowlers_to_analyze = [
    "R Vinay Kumar", "Mohammad Siraj", "KV Sharma", "Akash Deep", "GJ Maxwell", 
    "MK Lomror", "Mayank Dagar", "RJW Topley", "C Green", "AS Joseph", 
    "Yash Dayal", "TK Curran", "LH Ferguson"
]

filtered_bowlers = [bowler for bowler in bowlers_to_analyze if bowler in unique_bowler_names]

# Recalculating the metrics for the corrected list of bowlers
bowling_analysis = df[df['bowler'].isin(filtered_bowlers)].groupby('bowler').agg(
    Matches_Played=pd.NamedAgg(column='matchId', aggfunc=pd.Series.nunique),
    Balls_Bowled=pd.NamedAgg(column='over_ball', aggfunc='count'),
    Total_Runs_Conceded=pd.NamedAgg(column='Total_Runs_Conceded', aggfunc='sum')
)

# Calculating Bowling Economy Rate
bowling_analysis['Bowling_Economy_Rate'] = (bowling_analysis['Total_Runs_Conceded'] / bowling_analysis['Balls_Bowled']) * 6

bowling_analysis
```

```python
# Defining the list of bowlers to analyze for wickets taken
bowlers_to_analyze_wickets = [
    "R Vinay Kumar", "Mohammad Siraj", "KV Sharma", "Akash Deep", "GJ Maxwell", 
    "MK Lomror", "Mayank Dagar", "RJW Topley", "C Green", "AS Joseph", 
    "Yash Dayal", "TK Curran", "LH Ferguson"
]

# Filtering the dataset for these bowlers
df_wickets = df[df['bowler'].isin(bowlers_to_analyze_wickets)]

# For wickets, we consider 'isBowlersWkt' column; a value of 1 indicates a wicket taken by the bowler
# Calculating wickets taken, not including run outs
# Strike rate = balls bowled / wickets taken
wickets_analysis = df_wickets.groupby('bowler').agg(
    Matches_Played=pd.NamedAgg(column='matchId', aggfunc=pd.Series.nunique),
    Balls_Bowled=pd.NamedAgg(column='over_ball', aggfunc='count'),
    Wickets_Taken=pd.NamedAgg(column='isBowlersWkt', aggfunc='sum')
)

# Calculating the Strike Rate
wickets_analysis['Strike_Rate'] = wickets_analysis['Balls_Bowled'] / wickets_analysis['Wickets_Taken']

# Adjusting for bowlers who might be listed under slightly different names or not have data on wickets
# This ensures we only display those bowlers for whom we have data
wickets_analysis_final = wickets_analysis.reset_index().replace({
    "bowler": {
        "Mohammad Siraj": "Mohammed Siraj",
        "KV Sharma": "Karn Sharma",
        "GJ Maxwell": "Glenn Maxwell",
        "MK Lomror": "Mahipal Lomror",
        "RJW Topley": "Reece Topley",
        "C Green": "Cameron Green",
        "AS Joseph": "Alzarri Joseph",
        "TK Curran": "Tom Curran",
        "LH Ferguson": "Lockie Ferguson"
    }
}).set_index('bowler')

wickets_analysis_final
```

```python
# To calculate the percentage of dot balls bowled by each bowler,
# we first identify deliveries where no runs were scored by the batsman (batsman_runs = 0),
# and then calculate the percentage of such deliveries out of the total deliveries bowled by each bowler.

# Identifying dot balls (where batsman_runs = 0)
df_wickets['isDotBall'] = df_wickets['batsman_runs'].apply(lambda x: 1 if x == 0 else 0)

# Grouping by bowler to calculate the total number of dot balls and then calculate the percentage
dot_balls_analysis = df_wickets.groupby('bowler').agg(
    Total_Balls_Bowled=pd.NamedAgg(column='over_ball', aggfunc='count'),
    Dot_Balls=pd.NamedAgg(column='isDotBall', aggfunc='sum')
)

dot_balls_analysis['Dot_Ball_Percentage'] = (dot_balls_analysis['Dot_Balls'] / dot_balls_analysis['Total_Balls_Bowled']) * 100

# Adjusting for bowlers who might be listed under slightly different names or not have data on dot balls
# This ensures we only display those bowlers for whom we have dot ball data
dot_balls_analysis_final = dot_balls_analysis.reset_index().replace({
    "bowler": {
        "Mohammad Siraj": "Mohammed Siraj",
        "KV Sharma": "Karn Sharma",
        "GJ Maxwell": "Glenn Maxwell",
        "MK Lomror": "Mahipal Lomror",
        "RJW Topley": "Reece Topley",
        "C Green": "Cameron Green",
        "AS Joseph": "Alzarri Joseph",
        "TK Curran": "Tom Curran",
        "LH Ferguson": "Lockie Ferguson"
    }
}).set_index('bowler')

dot_balls_analysis_final[['Dot_Balls', 'Dot_Ball_Percentage']]
```

```python
# Calculating the number of 'No Balls', number of 6s, and number of 4s conceded by each bowler

# Number of No Balls conceded
df_wickets['isNoBall'] = df_wickets['isNoBall'].fillna(0)
no_balls_analysis = df_wickets.groupby('bowler').agg(
    No_Balls_Conceded=pd.NamedAgg(column='isNoBall', aggfunc='sum')
)

# Number of 6s conceded
df_wickets['isSix'] = df_wickets['batsman_runs'].apply(lambda x: 1 if x == 6 else 0)
sixes_conceded_analysis = df_wickets.groupby('bowler').agg(
    Sixes_Conceded=pd.NamedAgg(column='isSix', aggfunc='sum')
)

# Number of 4s conceded
df_wickets['isFour'] = df_wickets['batsman_runs'].apply(lambda x: 1 if x == 4 else 0)
fours_conceded_analysis = df_wickets.groupby('bowler').agg(
    Fours_Conceded=pd.NamedAgg(column='isFour', aggfunc='sum')
)

# Merging the analysis dataframes
merged_analysis = pd.concat([no_balls_analysis, sixes_conceded_analysis, fours_conceded_analysis], axis=1)

# Adjusting for bowlers who might be listed under slightly different names
merged_analysis_final = merged_analysis.reset_index().replace({
    "bowler": {
        "Mohammad Siraj": "Mohammed Siraj",
        "KV Sharma": "Karn Sharma",
        "GJ Maxwell": "Glenn Maxwell",
        "MK Lomror": "Mahipal Lomror",
        "RJW Topley": "Reece Topley",
        "C Green": "Cameron Green",
        "AS Joseph": "Alzarri Joseph",
        "TK Curran": "Tom Curran",
        "LH Ferguson": "Lockie Ferguson"
    }
}).set_index('bowler')

merged_analysis_final
```

```python
# Correcting the approach to calculate the percentages of runs conceded in 4s and 6s
# We will use the 'Total_Runs_Conceded' from the 'bowling_analysis' DataFrame that we calculated earlier

# Merging the total runs conceded with the no balls, sixes, and fours conceded analysis
analysis_with_total_runs = merged_analysis.join(bowling_analysis['Total_Runs_Conceded'])

# Calculating runs conceded in 4s and 6s
analysis_with_total_runs['Runs_Conceded_in_4s'] = analysis_with_total_runs['Fours_Conceded'] * 4
analysis_with_total_runs['Runs_Conceded_in_6s'] = analysis_with_total_runs['Sixes_Conceded'] * 6

# Calculating the percentages of total runs conceded in 4s and 6s
analysis_with_total_runs['Percentage_of_Runs_in_4s'] = (analysis_with_total_runs['Runs_Conceded_in_4s'] / analysis_with_total_runs['Total_Runs_Conceded']) * 100
analysis_with_total_runs['Percentage_of_Runs_in_6s'] = (analysis_with_total_runs['Runs_Conceded_in_6s'] / analysis_with_total_runs['Total_Runs_Conceded']) * 100

# Extracting the relevant columns for the final analysis
final_percentages = analysis_with_total_runs[['Total_Runs_Conceded', 'Runs_Conceded_in_4s', 'Percentage_of_Runs_in_4s', 'Runs_Conceded_in_6s', 'Percentage_of_Runs_in_6s']]

final_percentages
```

#Separate the data into two parts based on the inning – one dataset for the 1st inning and another for the 2nd inning

```python
# Check unique values in the 'inning' column
print(df['inning'].unique())
```

```python
# Filter the dataset for the 1st and 2nd innings
first_inning_data = df[df['inning'] == 1]
second_inning_data = df[df['inning'] == 2]
```

```python

# List of specified bowlers
specified_bowlers = [
    "R Vinay Kumar", "Mohammad Siraj", "KV Sharma", "Akash Deep", "GJ Maxwell",
    "MK Lomror", "Mayank Dagar", "RJW Topley", "C Green", "AS Joseph",
    "Yash Dayal", "TK Curran", "LH Ferguson"
]

def calculate_bowling_metrics_for_specified_bowlers(df, bowlers):
    # Filter for specified bowlers
    df = df[df['bowler'].isin(bowlers)]
    
    # Filter deliveries for dot balls calculation (excluding wides and no balls)
    valid_deliveries = df[(df['extras'] == 0) | (df['isWide'].isna() & df['isNoBall'].isna())]
    
    # Dot Balls
    dot_balls = valid_deliveries[valid_deliveries['batsman_runs'] == 0].groupby('bowler').size()
    total_balls = valid_deliveries.groupby('bowler').size()
    dot_ball_percentage = (dot_balls / total_balls * 100).rename("DotBallPercentage")
    
    # Wickets Taken (excluding run outs and other non-bowler dismissals)
    wickets = df[df['dismissal_kind'].isin(['caught', 'bowled', 'lbw', 'stumped', 'caught and bowled', 'hit wicket'])]
    wickets_taken = wickets.groupby('bowler').size().rename("WicketsTaken")
    
    # Economy Rate = Total Runs Conceded / Overs Bowled (where 1 over = 6 balls)
    runs_conceded = df.groupby('bowler')['batsman_runs'].sum()
    overs_bowled = total_balls / 6
    economy_rate = (runs_conceded / overs_bowled).rename("EconomyRate")
    
    # Merging the metrics
    bowling_metrics = pd.concat([dot_ball_percentage, economy_rate, wickets_taken], axis=1, sort=False).fillna(0)
    
    return bowling_metrics

# Calculating metrics for the 1st and 2nd innings for the specified bowlers
bowling_metrics_1st_inning_specified = calculate_bowling_metrics_for_specified_bowlers(first_inning_data, specified_bowlers)
bowling_metrics_2nd_inning_specified = calculate_bowling_metrics_for_specified_bowlers(second_inning_data, specified_bowlers)

# Display the metrics for the specified bowlers
print("1st Innings Bowling Metrics:")
print(bowling_metrics_1st_inning_specified)
print("\n2nd Innings Bowling Metrics:")
print(bowling_metrics_2nd_inning_specified)
```

```python
## Power Play 
# Assuming 'deliveries_df_csv' is your DataFrame containing detailed delivery information
# Including at least 'bowler', 'over', 'total_runs', 'is_wicket' columns

# Filter for power play overs (1-6)
power_play_deliveries = df[df['over'].isin(range(1, 6))]

# Further filter for the specified bowlers
specified_bowlers = ['R Vinay Kumar', 'Mohammad Siraj', 'KV Sharma', 'Akash Deep', 'GJ Maxwell',
                     'MK Lomror', 'Mayank Dagar', 'RJW Topley', 'C Green', 'AS Joseph',
                     'Yash Dayal', 'TK Curran', 'LH Ferguson']
power_play_deliveries_specified = power_play_deliveries[power_play_deliveries['bowler'].isin(specified_bowlers)]

# Calculate performance metrics
# Example: Economy Rate = Total Runs Conceded / Overs Bowled
# Assuming 'is_wicket' is 1 for wicket-taking deliveries and 'total_runs' includes all runs from the delivery
metrics = power_play_deliveries_specified.groupby('bowler').agg(
    Total_Runs_Conceded=pd.NamedAgg(column='batsman_runs', aggfunc='sum'),
    Balls_Bowled=pd.NamedAgg(column='ball', aggfunc='count'),
    Wickets_Taken=pd.NamedAgg(column='isBowlersWkt', aggfunc='sum')
).reset_index()

# Calculate Economy Rate and Strike Rate
metrics['Economy_Rate'] = metrics['Total_Runs_Conceded'] / (metrics['Balls_Bowled'] / 6)
metrics['Strike_Rate'] = metrics['Balls_Bowled'] / metrics['Wickets_Taken']

# Displaying the metrics
print(metrics[['bowler', 'Economy_Rate', 'Strike_Rate', 'Wickets_Taken']])
```

```python

# Assuming 'deliveries_df' is your DataFrame

# Specified bowlers list
specified_bowlers = ['R Vinay Kumar', 'Mohammad Siraj', 'KV Sharma', 'Akash Deep', 'GJ Maxwell',
                     'MK Lomror', 'Mayank Dagar', 'RJW Topley', 'C Green', 'AS Joseph',
                     'Yash Dayal', 'TK Curran', 'LH Ferguson']

# Example: Filtering for Powerplay overs in 1st and 2nd innings and calculating economy rate
for inning in [1, 2]:
    powerplay_deliveries = df[
        (df['over'].isin(range(1, 7))) & 
        (df['inning'] == inning) & 
        (df['bowler'].isin(specified_bowlers))
    ]

    # Calculate total runs conceded and balls bowled by bowler
    metrics = powerplay_deliveries.groupby('bowler').agg(
        Total_Runs_Conceded=pd.NamedAgg(column='total_runs', aggfunc='sum'),
        Balls_Bowled=pd.NamedAgg(column='ball', aggfunc='count')
    ).reset_index()

    # Calculate Economy Rate
    metrics['Economy_Rate'] = metrics['Total_Runs_Conceded'] / (metrics['Balls_Bowled'] / 6)

    # Visualization for each inning
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Economy_Rate', y='bowler', data=metrics.sort_values(by='Economy_Rate'), palette='Set2')
    plt.title(f'Powerplay Economy Rate - Inning {inning}')
    plt.xlabel('Economy Rate')
    plt.ylabel('Bowler')
    plt.show()
```
